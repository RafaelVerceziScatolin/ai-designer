```python
import sys
from pathlib import Path
sys.path.append(str(Path.home() / "orguel_ml"))
from orguel_ml.ocr import character_set, width_factors

# Output directory
#output_folder = r"C:\Users\rafael.scatolin\OneDrive - Orguel\ai_designer_database\dxf files\9-characters\individual"
output_folder = r"C:\Users\Rafael\Desktop\training_dataset\dxf files\9-characters\individual"
```

```python
import os
import ezdxf

def GenerateDXF(font_name, font_file, character, width_factor=1):
    doc = ezdxf.new()
    modelSpace = doc.modelspace()
    styles = doc.styles
    
    if font_name not in styles: styles.new(font_name, dxfattribs={"font": font_file})
    
    text = modelSpace.add_text(
        character,
        dxfattribs={
            "height": 1,
            "style": font_name,
            "rotation": 0,
            "width": width_factor
            }
        )
    text.dxf.insert = (0, 0)
    
    suffix = f"_ascii{ord(character)}"
    suffix += f"_width{width_factor:.2f}".replace('.', '_')
    savePath = os.path.join(output_folder, f"{font_name}{suffix}.dxf")
    doc.saveas(savePath)
    print(f"Saved: {savePath}")
    
```

```python
for fontName, fontFile in character_set["font"]["type"].items():
    for group, characters in character_set["characters"]["type"].items():
        for character in characters:
            for widthFactor in width_factors:
                GenerateDXF(fontName, fontFile, character, widthFactor)
```
