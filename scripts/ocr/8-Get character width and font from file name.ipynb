```python
import os
import re
import pickle
from pathlib import Path

with open(str(Path.home()/"orguel_ml"/"scripts"/"ocr"/"character_positions.pkl"), "rb") as f:
    characterPositions = pickle.load(f)

def GetWidth(filepath):
    filename = os.path.basename(filepath)  # Remove folder paths
    # Look for width pattern either after "__" or after "_width"
    match = re.search(r'(?:__|_width)(\d+)_(\d+)(?=\.dxf$)', filename)
    if match:
        whole, decimal = match.groups()
        return float(f"{whole}.{decimal}")
    else:
        return None

def GetFont(filepath):
    filename = os.path.basename(filepath)  # Remove folder paths
    if "_ascii" in filename: font = filename.split("_ascii")[0]
    elif "_width" in filename: font = filename.split("_width")[0]
    elif "__" in filename: font = filename.split("__")[0].split("_")[-1] 
    return font
```

```python
# Apply it to your dataframe
characterPositions['width'] = characterPositions['file'].apply(GetWidth)
characterPositions['font'] = characterPositions['file'].apply(GetFont)
```
